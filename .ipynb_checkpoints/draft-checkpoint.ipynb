```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.geocoders import Nominatim
import geopandas as gpd
import folium
folder = '/home/hiba/projet/transactions-ser'
```

```python
csv_files = [f for f in os.listdir(folder) if f.endswith('.csv')]
```

```python
df = pd.read_csv(os.path.join(folder, csv_files[0]),delimiter=';')
# Get a list of the columns in the first CSV file
common_columns = list(df.columns)
```

```python
for file in csv_files[1:]:
    df_temp = pd.read_csv(os.path.join(folder, file), delimiter=';')
    # Keep only the columns that are present in the first CSV file
    df_temp = df_temp[common_columns]
    df = df.append(df_temp, ignore_index=False)
df.to_csv('merged.csv')
```

```python
df = df.replace({'None': None, 'nan': float('nan')})
df
```

```python
df.info()
```

### Nettoyage du dataFrame

```python
df = df.dropna(thresh=len(df) * 0.25, axis=1)
df = df.dropna(axis=0)
df = df.drop_duplicates()
df['code_type_local'] = pd.to_numeric(df['code_type_local'], errors='coerce', downcast='integer')
df['date_mutation'] = pd.to_datetime(df['date_mutation'])
df['année'] = df['date_mutation'].dt.year
df = df.drop(['code_postal','code_commune', 'section_prefixe','nom_commune','code_departement','id_parcelle','adresse_code_voie','adresse_nom_voie'], axis=1)
```

```python
df = df.drop(['id_mutation'], axis=1)
```

```python
df["code_nature_culture"] = df["code_nature_culture"].replace({'S':1, 'AG':2, 'T':3, 'J':4, 'L':5,'AB':6})
```

```python
df["nature_mutation"] = df["nature_mutation"].replace({'Vente':1, 'Adjudication':2, "Vente en l'état futur d'achèvement":3})
```

```python
df = df.drop(df[df["type_local"] == 'Local industriel. commercial ou assimilé'].index)
```

## Régression linéaire :

```python
X = df[['code_type_local','surface_reelle_bati','surface_terrain','nombre_pieces_principales']]
y = df['valeur_fonciere']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
accuracy = lm.score(X_test, y_test)
"Accuracy: {}%".format(int(round(accuracy * 100)))
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])

coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
#plt.ticklabel_format(style = 'plain')
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(y_test,predictions)
ax.set_xlim(0,2000000 )
ax.set_ylim(0,2000000 )
```

```python
sns.distplot((y_test-predictions),bins=30)
ax = plt.gca()
ax.ticklabel_format(style='plain')

# Show the plot
plt.show()
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))

```
